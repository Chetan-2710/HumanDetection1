```python
# Paths to the folder and CSV file
folder_path = r"C:\Users\kolak\3_1_project\DataSet1\ntut_drone_train\ntut_drone_train\Drone_023\vott-csv-export"  # Replace with your folder path
input_csv = r"C:\Users\kolak\3_1_project\DataSet1\ntut_drone_train\ntut_drone_train\Drone_023\Drone_023-export.csv"  # Replace with your CSV file path

# Read the CSV into a DataFrame
df = pd.read_csv(input_csv)

# Function to extract data for a given filename
def extract(filename_no_ext):
    # Check if the filename exists in the CSV
    if filename_no_ext in df['image'].values:  # Adjust 'image' to match the filename column in your CSV
        # Extract all rows corresponding to the image
        image_rows = df[df['image'] == filename_no_ext]  # Filter rows matching the image name
        image_path = os.path.join(folder_path, filename_no_ext)
        # Read the image using OpenCV to get dimensions
        image = cv2.imread(image_path)
        temp=[]
        results = []  # To store the results
        for _, row in image_rows.iterrows():
            if image is not None:
                height, width = image.shape[:2]
            else:
                height, width = None, None
            label = row['label']  
            xmin = row['xmin']   
            xmax = row['xmax']  
            ymin = row['ymin']   
            ymax = row['ymax']
            if label.lower().startswith("block"):
                temp.append([xmin,xmax,ymin,ymax])
        print(temp)
        for _, row in image_rows.iterrows():  # Iterate over matching rows
            if image is not None:
                height, width = image.shape[:2]  # Extract image dimensions
            else:
                height, width = None, None  # Handle missing or unreadable images
            # Extract values from the row
            label = row['label']  # Adjust 'label' to match the actual column name in your CSV
            xmin = row['xmin']   # Adjust 'xmin' to match your CSV column
            xmax = row['xmax']   # Adjust 'xmax' to match your CSV column
            ymin = row['ymin']   # Adjust 'ymin' to match your CSV column
            ymax = row['ymax']   # Adjust 'ymax' to match your CSV column

            # If the label is not "block", change it to "human"
            if [xmin,xmax,ymin,ymax] not in temp and not label.startswith("id_") and height!=None and width!=None:
                label = "human"
                # Append the data to results
                results.append([filename_no_ext, width, height, label, xmin, xmax, ymin, ymax])
                # print(count[0])
                # count[0]+=1
            else:
                rejected.append(filename_no_ext)
        
        return results
    else:
        print(f"Image not found in CSV: {filename_no_ext}")
        # return []  # Return an empty list if no matching rows are found
```

```python
df = pd.DataFrame(data,columns = ['filename','width','height','name','xmin','xmax','ymin','ymax'])
```

```python
# center x, center y
df['center_x'] = ((df['xmax']+df['xmin'])/2)/df['width']
df['center_y'] = ((df['ymax']+df['ymin'])/2)/df['height']
# w 
df['w'] = (df['xmax']-df['xmin'])/df['width']
# h 
df['h'] = (df['ymax']-df['ymin'])/df['height']
```

```python
# img_train = tuple(df['filename'].unique().to_pandas().sample(frac=1.0))
img_train = tuple(pd.DataFrame(df['filename'].unique(), columns=['filename']).sample(frac=1.0)['filename'])
```

```python
train_df = df[df['filename'].isin(img_train)]
```

```python
# Encode labels
def label_encoding(x):
    labels = {'human': 0}
    return labels.get(x, -1)
```

```python
train_df['id'] = train_df['name'].apply(label_encoding)
print("encoding done")
```

```python
train_folder = r"C:\Users\kolak\3_1_project\DataSet1\ntut_drone_train\ntut_drone_train\Drone_023\train2"
cols = ['filename', 'id', 'center_x', 'center_y', 'w', 'h']
groupby_obj_train = train_df[cols].groupby('filename')
```

```python
count=[1]
# Save each image in train/test folder and respective labels in .txt
def save_data(filename, folder_path, group_obj):
    # Save the labels
    text_filename = os.path.join(folder_path, os.path.splitext(filename)[0]+ '.txt')
    group_obj.get_group(filename).set_index('filename').to_csv(text_filename, sep=' ', index=False, header=False)
    print(count[0])
    count[0]+=1
```

```python
filename_series = pd.Series(groupby_obj_train.groups.keys())
```

```python
filename_series.apply(save_data, args=(train_folder, groupby_obj_train))
```
